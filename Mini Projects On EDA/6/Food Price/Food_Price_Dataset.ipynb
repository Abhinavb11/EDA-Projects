# About Data

## Overview

The dataset consists of several features that provide valuable information about food prices. The date feature represents the specific day on which the price was recorded, making it useful for identifying trends and seasonal variations. Geographic details are captured by admin1 and admin2, which represent the state or region and city or district, respectively. The market column specifies the particular market where the price was recorded, adding another layer of geographic granularity. Additionally, latitude and longitude provide precise spatial coordinates for the markets.

## Features

1. date: The date on which the price of the commodity was recorded. It helps track trends or analyze seasonal variations in prices.
2. admin1: The first administrative division, such as a state or region, providing geographic context for the data.
3. admin2: The second administrative division, such as a city or district, offering more granular geographic detail.
4. market: The specific market where the data was collected. Useful for market-specific analyses.
5. latitude: The geographical latitude of the market, providing spatial location data.
6. longitude: The geographical longitude of the market, complementing the latitude for spatial analysis.
7. category: The broader category of the commodity, such as grains, oil and fats, etc. It groups commodities for aggregated analysis.
8. commodity: The specific item being priced, such as rice, wheat, or sugar. Critical for understanding price variation across different items.
9. unit: The unit of measurement for the commodity price, e.g., KG. Ensures clarity in interpreting the price data.
10. priceflag: Indicates whether the price is an actual recorded value or an estimated one. It’s essential for data quality checks.
11. pricetype: Specifies whether the price is retail, wholesale, or another type. Useful for comparing price types.
12. currency: The currency in which the price is recorded (e.g., INR). It’s important for currency-specific analysis or conversion.

## Target Variable

* price: The price of the commodity in the local currency. It represents the value to be predicted or analyzed in most price-related tasks.

```python
import pandas as pd

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['unit'].quantile(0.25)
Q3 = df['unit'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df = df[(df['unit'] >= lower_bound) & (df['unit'] <= upper_bound)]
```

- **unit**: Measurement unit (e.g., KG, Liter).
- **priceflag**: Indicates if the price is actual or estimated.
- **pricetype**: Type of price (Retail, Wholesale, etc.).
- **currency**: Currency of the recorded price.
- **price**: Recorded price of the commodity.
- **usdprice**: Price converted to USD for standardization.
